```python
import pandas as pd
```

```python
winmag_url = 'https://gist.githubusercontent.com/clairehq/79acab35be50eaf1c383948ed3fd1129/raw/407a02139ae1e134992b90b4b2b8c329b3d73a6a/winemag-data-130k-v2.csv'
```

```python
df = pd.read_csv(winmag_url, index_col=0)
display(df)
```

```python
# Número de variables (columnas) en el dataframe
num_variables = df.shape[1]
print(f"El dataframe tiene {num_variables} variables.")
```

```python
# Variables con valores perdidos
valores_perdidos = df.isnull().sum()
print("Variables con valores perdidos:")
print(valores_perdidos[valores_perdidos > 0])
```

```python
# Variables numéricas
variables_numericas = df.select_dtypes(include=['number']).columns
print("Variables numéricas:")
print(variables_numericas)
```

```python
# Variables cualitativas
variables_cualitativas = df.select_dtypes(include=['object']).columns
print("Variables cualitativas:")
print(variables_cualitativas)
```

```python
# Filtrar el dataframe para incluir solo vinos europeos
df_paises = df['country'].unique()
# Mostrar el dataframe de vinos europeos
print(df_paises)
```

```python
# Definir la lista de países europeos
paises_europeos = [
    'France', 'Italy', 'Spain', 'Portugal', 'Germany', 'Austria',
    'Hungary', 'Greece', 'Switzerland', 'Belgium', 'Netherlands',
    'Czech Republic', 'Slovakia', 'Croatia', 'Serbia', 'Bulgaria',
    'Romania', 'Moldova', 'Luxembourg', 'Malta'
]
```

```python
# Filtrar el dataframe para incluir solo vinos europeos
df_europeos = df[df['country'].isin(paises_europeos)]
```

```python
display(df_europeos)
```

```python
# Calcular el menor, mayor y promedio del precio de la botella por país
precios_por_pais = df.groupby('country')['price'].agg(['min', 'max', 'mean']).reset_index()

# Mostrar los resultados
print("Precios por país (mínimo, máximo, promedio):")
print(precios_por_pais)

# Identificar el país de la botella de menor precio
pais_menor_precio = df.loc[df['price'].idxmin()]['country']
print(f"El país de la botella de menor precio es: {pais_menor_precio}")
```

```python
# Contar cuántos vinos tienen *aroma a fresa* en la descripción en el dataframe original
vinos_fresa = df[df['description'].str.contains('strawberry', case=False, na=False)]
num_vinos_fresa = len(vinos_fresa)

print(f"Hay {num_vinos_fresa} vinos con aroma a fresa.")
```

```python
# Contar cuántas designaciones diferentes hay
num_designaciones_unicas = df['designation'].nunique()
print(f"Hay {num_designaciones_unicas} designaciones diferentes.")

# Identificar la designación más repetida
designacion_mas_repetida = df['designation'].value_counts().idxmax()
# Contar cuántas veces se repite la designación más repetida
repeticiones_designacion = df['designation'].value_counts().max()

print(f"La designación más repetida es '{designacion_mas_repetida}', y se repite {repeticiones_designacion} veces.")
```

```python
# Crear un dataframe con variedad, país y precio para vinos con un costo menor a $20
df_vinos_baratos = df[df['price'] < 20][['variety', 'country', 'price']]

# Mostrar las primeras filas del dataframe resultante
display(df_vinos_baratos)
```

```python
# Contar cuántos vinos diferentes de la variedad *Pinot Noir* hay por cada país
vinos_pinot_noir = df[df['variety'] == 'Pinot Noir']
conteo_pinot_noir_por_pais = vinos_pinot_noir.groupby('country').size()

# Mostrar los resultados
print("Cantidad de vinos Pinot Noir por país:")
print(conteo_pinot_noir_por_pais)
```
